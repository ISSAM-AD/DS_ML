** I will create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. Returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
bank_stocks.head(3)
```

```python
returns_df = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
tickers
```

```python
df1 = bank_stocks.xs(key="Close",level=1,axis=1)
df1
```

```python
df1["BAC"]
```

```python
df1["BAC"].pct_change()

# 47.08 - 100%
# 46.58 - X

# X is 98.94% or 0.9894
# hence difference is -0.01062
```

```python
tickers
```

```python
for tick in tickers:
    returns_df[tick+" Return"] = df1[tick].pct_change()
```

```python
returns_df.head()
```
